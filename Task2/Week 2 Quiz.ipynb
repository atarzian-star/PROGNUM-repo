```python
# 2.12
from math import atan2, degrees

# Asks for input
x = float(input("Please enter a value for x = "))
y = float(input("Please enter a value for y = "))

# Calculates the angle using the function atan2, then transforms to degrees.
angle = atan2(y,x)
degrees = degrees(angle)

print("The angle in degrees is", degrees)
print("The angle in radians is", angle)

# Choosing which quadrant we are in
if (x >= 0 and y >= 0):
    print("We are in the first quadrant.")
elif (x < 0 and y >= 0):
    print("We are in the second quadrant.")
elif (x < 0  and y < 0):
    print("We are in the third quadrant.")
elif (x >= 0 and y < 0):
    print("We are in the fourth quadrant.")
```

```python
# 2.14
# List
X = [1.36,  1.46,  1.41,  2.12,  0.42,  0.98,  0.45,  1.97, -0.24, 2.18]

# Adding a new element in the list and removing 2.12
X.append(0.51)
X.remove(2.12)

# The value at index 6 is 1.97
# The index for value 1.46 is 1

# Method 1 of slicing
X_slice1 = X[2:8:2]
print(f"X slice is = {X_slice1}")

# Method 2 of slicing, using slice() function
indices = slice(2,8,2)
X_slice2 = X[indices]
print(f"X slice is = {X_slice2}")

# Making the sum and printing it
Sum = 0

for i in X_slice1:
    Sum = Sum + i
    
print("Sum of all values in the slice is:", Sum)
```

```python
# 2.15
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

# Making sets out of the lists
set1 = set(night1)
set2 = set(night2)

# Making the intersections
inters = set1.intersection(set2)
inters_list = list(inters)

# Printing the required stuff
print("The common elements are: ", inters_list)

print("Is 5 in the intersection list?", 5 in inters_list)
print("Is 5 in set(night1)?", 5 in set1)
```
