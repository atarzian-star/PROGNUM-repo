```python
# 3.2

from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

esc_vel = []

# Calculating the escape velocities and adding them to a list
for i in range(len(masses)):
    v = (2*G*masses[i] / radii[i])**0.5
    esc_vel.append(v)
    print(f"The mass is {masses[i]:.2e} kg")    # printing the masses

print()
print("The escape velocities in m/s are:", esc_vel)
print()

# Creating a dictionary
dictionary = dict(zip(names, esc_vel))

print("The dicitonary is:")
print(dictionary)
```

```python
# 3.7

import numpy as np

# Creating the function for angular size
def angular_size(d, D):
    a_rad = 2 * np.arctan(d / (2 * D))
    a_degrees = np.degrees(a_rad)
    return a_degrees

# Asking for user input for d and D
d = eval(input("What is the object size d? "))
D = eval(input("What is the distance D? "))

print("The angular size is:", angular_size(d, D))
```

```python
# 3.10

from math import pi

# Creating the function to find the area and volume of a cylinder
def cylinder(r, h):
    A = 2*pi*(r**2) + 2*pi*r*h
    V = pi*h*(r**2)
    return A, V

# Asking for user input for r and h
r = eval(input("What is the radius in m? "))
h = eval(input("What is the heightin m? "))

# I am printing this for myself
print("The area and volume are:", cylinder(r, h))

# Getting the values out of the tuple
area, volume = cylinder(r, h)
ratio = area / volume 

print("The ratio between the area and the volume is", ratio)
```

```python
# 3.17

import numpy as np

# Making the function to calculate the standard deviation
def stdev(y):
    
    std = ((np.sum((y - np.average(y))**2))/y.size)**0.5
    
    return std 

N = 5
y = np.random.normal(0, 2, N)

print(stdev(y), "\n")
print(np.std(y), "\n")
# They are the same now
```

```python
# 3.18

import numpy as np

# Transforming the lists into ndarrays
x = [18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]
y = [2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158]

d = np.array(x)
v = np.array(y)

# Finding the average for the formula
md = d.mean()
mv = v.mean()

# Finding a and b
b1 = 0
b2 = 0

for i in range(0, 13):
    b1 += (d[i] - md)*(v[i] - mv) 
    b2 += (d[i] - md)**2     

b = b1 / b2
a = mv - (b*md)

# Printing the results
print(a)
print("The slope is:", b, "km/s/Mpc")
```
